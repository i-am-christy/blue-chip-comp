# Categorical Encoding

```python
#encoding the training data
encoded = pd.get_dummies(cat_cols, prefix="cat")
```

```python
encoded
```

```python
#encoding the test data
test_encoded = pd.get_dummies(test_cat_cols, prefix="cat")
```

```python
#merging the encoded and the numeric data together fr the training adatframe
x_train = num_cols.join(encoded)
x_train.set_index("id", inplace=True)
```

```python
#merging the encoded and the numeric data together fr the test dataframe
test_df = test_num_cols.join(test_encoded)
test_df.set_index("id", inplace=True)
```

```python
test_df.head()
```

```python
x_train.head()
```

```python
#duplicated = x_train.columns[x_train.columns.duplicated()]
x_train = x_train.T.groupby(level=0).first().T
x_train.columns
```

# Feature Scaling

```python
#importing the robust scaler
from sklearn.preprocessing import RobustScaler 
```

```python
#initializing the scaler
robust_scaler = RobustScaler()
```

```python
#sacling the data
x_train = robust_scaler.fit_transform(x_train)
```

```python
#observing the scaled data
x_train.shape
```

# spliting the data

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the data into training and testsing data
x_train, X_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, stratify=y_train)
```

# Model Training

```python
#importing the model
from sklearn.ensemble import GradientBoostingClassifier
```

```python
#instatiate the model
GBClassifier = GradientBoostingClassifier()
```

```python
#fit our data
GBClassifier.fit(x_train, y_train)
```

```python
#use the model to get a prediction
GB_ypred = GBClassifier.predict(X_test)
```

# Model Evaluation

```python
#importing evaluation library
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

#evaluating our model accuracy
accuracy_score(y_test, GB_ypred)
```

```python
#plotting the confusion matrix
from sklearn.metrics import plot_confusion_matrix

# Get confusion matrix for Gradient Boosting Classifier
plot_confusion_matrix(GBClassifier, X_test, y_test, normalize='true')
```

```python
#printing the classification report
print(classification_report(y_test, GB_ypred))
```

# Hyperparameter Tunning

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
#apecifying the hyperparameters in random forest to be tunned
parameters = {
    "criterion" : ["gini", "entropy"],
    "max_depth" : [1, 2, 4, 5, 10, 15],
    "min_samples_split" : [10, 20, 40, 50, 100, 150],
    "max_leaf_nodes" : [10, 15, 25, 50],
    "min_samples_leaf" : [100, 200, 300, 400, 500, 600],
    "n_estimators" : [0, 50, 100, 150, 200, 500],
    "max_samples" : [0, 0.1, 0.15, 0.2, 0.25, 0.3],
    "max_features" : [0, 1, 2, 3, 5, 7, "auto", "sqrt", "log2"]
}
```

```python
#creating the kfold object
kfold = StratifiedKFold(n_splits=5)
```

```python
#creating the grid search object
clf = GridSearchCV(random, parameters, scoring="accuracy", n_jobs=-1, cv=kfold)
```

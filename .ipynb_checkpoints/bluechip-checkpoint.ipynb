```python
#separating the explanatory and response variable
x_train = train_df.drop(['Attrition'], axis=1)
y_train = train_df["Attrition"]
x_train
#train_df.columns
```

```python
#separating numeric and categorical columns for the train df
num_cols = x_train.select_dtypes(["int", "float"])
cat_cols = x_train.select_dtypes("object")
```

```python
#separating the categorical data for the test df
test_num_cols = test_df.select_dtypes(["int", "float"])
test_cat_cols = test_df.select_dtypes("object")
```

```python
#detecting and removing outliers in the num_cols dataset
def plot_boxplot(df):
    for feature in df.columns:
        df.boxplot(column=[feature])
        plt.grid=(False)
        plt.show()

plot_boxplot(num_cols)
```

# Categorical Encoding

```python
#encoding the training data
encoded = pd.get_dummies(cat_cols, prefix="cat")
```

```python
encoded
```

```python
#encoding the test data
test_encoded = pd.get_dummies(test_cat_cols, prefix="cat")
```

```python
#merging the encoded and the numeric data together fr the training adatframe
x_train = num_cols.join(encoded)
x_train.set_index("id", inplace=True)
```

```python
#merging the encoded and the numeric data together fr the test dataframe
test_df = test_num_cols.join(test_encoded)
test_df.set_index("id", inplace=True)
```

# Model Training

```python
#importing the model
from sklearn.ensemble import GradientBoostingClassifier
```

```python
#instatiate the model
GBClassifier = GradientBoostingClassifier()
```

```python
#fit our data
GBClassifier.fit(x_train, y_train)
```

```python
#use the model to get a prediction
GB_ypred = GBClassifier.predict(test_df)
```

# Model Evaluation

```python
#evaluating our model
GBClassifier.score(test_df, GB_ypred)
```
